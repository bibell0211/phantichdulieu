```python
import pandas as pd
dulieuxettuyen = pd.read_csv("https://raw.githubusercontent.com/huynhhoc/phantichdulieu/main/Data/dulieuxettuyendaihoc.csv")
```

```python
dulieuxettuyen.head(10)
```

```python
dulieuxettuyen.tail(10)
```

```python
dulieuxettuyen['DT'] = dulieuxettuyen['DT'].fillna(0)
```

```python
dulieuxettuyen
```

```python
dulieuxettuyen['TBM1'] = (dulieuxettuyen['T1'] + dulieuxettuyen['L1'] + dulieuxettuyen['H1'] + dulieuxettuyen['S1'] + dulieuxettuyen['V1'] + dulieuxettuyen['X1'] + dulieuxettuyen['D1'] + dulieuxettuyen['N1']) /10
```

```python
dulieuxettuyen['TBM2'] = (dulieuxettuyen['T2'] + dulieuxettuyen['L2'] + dulieuxettuyen['H2'] + dulieuxettuyen['S2'] + dulieuxettuyen['V2'] + dulieuxettuyen['X2'] + dulieuxettuyen['D2'] + dulieuxettuyen['N2']) /10
```

```python
dulieuxettuyen['TBM3'] = (dulieuxettuyen['T3'] + dulieuxettuyen['L3'] + dulieuxettuyen['H3'] + dulieuxettuyen['S3'] + dulieuxettuyen['V3'] + dulieuxettuyen['X3'] + dulieuxettuyen['D3'] + dulieuxettuyen['N3']) /10
```

```python
dulieuxettuyen
```

```python
for i in range(len(dulieuxettuyen)):
  if dulieuxettuyen['TBM1'][i] < 5.0:
    dulieuxettuyen.loc[i, 'XL1'] = 'Y'
  elif dulieuxettuyen['TBM1'][i] < 6.5:
    dulieuxettuyen.loc[i, 'XL1'] = 'TB'
  elif dulieuxettuyen['TBM1'][i] < 8.0:
    dulieuxettuyen.loc[i, 'XL1'] = 'K'
  elif dulieuxettuyen['TBM1'][i] < 9.0:
    dulieuxettuyen.loc[i, 'XL1'] = 'G'
  else :
    dulieuxettuyen.loc[i, 'XL1'] = 'XS'
```

```python
dulieuxettuyen
```

```python
for i in range(len(dulieuxettuyen)):
  if dulieuxettuyen['TBM2'][i] < 5.0:
    dulieuxettuyen.loc[i, 'XL2'] = 'Y'
  elif dulieuxettuyen['TBM2'][i] < 6.5:
    dulieuxettuyen.loc[i, 'XL2'] = 'TB'
  elif dulieuxettuyen['TBM2'][i] < 8.0:
    dulieuxettuyen.loc[i, 'XL2'] = 'K'
  elif dulieuxettuyen['TBM2'][i] < 9.0:
    dulieuxettuyen.loc[i, 'XL2'] = 'G'
  else :
    dulieuxettuyen.loc[i, 'XL2'] = 'XS'
```

```python
for i in range(len(dulieuxettuyen)):
  if dulieuxettuyen['TBM3'][i] < 5.0:
    dulieuxettuyen.loc[i, 'XL3'] = 'Y'
  elif dulieuxettuyen['TBM3'][i] < 6.5:
    dulieuxettuyen.loc[i, 'XL3'] = 'TB'
  elif dulieuxettuyen['TBM3'][i] < 8.0:
    dulieuxettuyen.loc[i, 'XL3'] = 'K'
  elif dulieuxettuyen['TBM3'][i] < 9.0:
    dulieuxettuyen.loc[i, 'XL3'] = 'G'
  else :
    dulieuxettuyen.loc[i, 'XL3'] = 'XS'
```

```python
dulieuxettuyen
```

```python
dulieuxettuyen['US_TBM1'] = (dulieuxettuyen['TBM1'] - 0) / (10-0) * (4-0) + 0
```

```python
dulieuxettuyen['US_TBM2'] = (dulieuxettuyen['TBM2'] - 0) / (10-0) * (4-0) + 0
```

```python
dulieuxettuyen['US_TBM3'] = (dulieuxettuyen['TBM3'] - 0) / (10-0) * (4-0) + 0
```
